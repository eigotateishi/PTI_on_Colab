```python
l_src_theme = []

for i in range(len(l_tar_ISOs)):

    # Country directory
    iso_temp = l_tar_ISOs[i]
    src_ISO_dir = os.path.join(os.getcwd(), 'data', iso_temp)

    if not os.path.exists(src_ISO_dir):
      os.mkdir(src_ISO_dir)

    # Theme directories
    src_WP = os.path.join(src_ISO_dir, 'WorldPop')
    src_LH = os.path.join(src_ISO_dir, 'LivelihoodZones')
    src_GRID3 = os.path.join(src_ISO_dir, 'GRID3')
    src_CNF = os.path.join(src_ISO_dir, 'Conflict')
    src_AGRI = os.path.join(src_ISO_dir, 'Agriculture')
    src_ADM = os.path.join(src_ISO_dir, 'ADM')

    l_src_theme = [src_WP, src_LH, src_GRID3, src_CNF, src_AGRI, src_ADM]


    print('Process : {}'.format(iso_temp))

    for t in l_src_theme:
      print(t)

      if not os.path.exists(t):
        os.mkdir(t)

      for s in l_src_sub:

        src2generate = os.path.join(t, s)
        print(src2generate)

        if not os.path.exists(src2generate):
          os.mkdir(src2generate)

```
